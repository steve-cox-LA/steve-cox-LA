```python
# we now graph the peak-to-peak output voltage over the range of frequencies

import matplotlib.pyplot as plt   
import numpy as np

freq = np.array([1, 10, 100, 1000])      # Hz
VoutPP = np.array([1, 0.82, 0.15, 0.02])  # Volts, read from oscilloscope.

plt.semilogx(freq, VoutPP, 'ko-')
plt.grid('on')
plt.xlabel('frequency, Hz')
plt.ylabel('VoutPP, Volts')
plt.title('Figure 1. LR Filter, $R=10\Omega$, $L=100\,mH$')
```

As in our RC notebooks we now seek a quantitative understanding of this Low Pass Filter phenomena.

To that end we draw a careful schematic, derive the Differential Equation for the output voltage, and solve it given our knowledge of the input voltage.

```python
import matplotlib.pyplot as plt   
import schemdraw
import schemdraw.elements as elm
d = schemdraw.Drawing()

R = d.add(elm.Inductor, label='$L$')
d.labelI(R, label='$I$', arrowofst=0.5, arrowlen=1.5, top=False)
node = d.add(elm.DOT)
node.add_label('$V_1$',loc = 'top') 

C = d.add(elm.Resistor, label='$R$', d='down')
d.labelI(C, label='$I$', arrowofst=1, arrowlen=1.5, top=True)

d.add(elm.LINE, d='left', tox = R.start)
d.add(elm.GND)
d.add(elm.SourceSin, d='up', label='$V_0$')

node = d.add(elm.DOT)
node.add_label('$V_0$',loc = 'top') 

d.draw()
```

The Inductor Law states that
$$
  V_0-V_1 = LI'
$$
while Ohm's Law provides
$$
  V_1 = IR
$$
Substitution of the second into the first reveals
$$
  V_0 - IR = LI'.
$$
We rearrange to find

$$
  \tau I' + I = V_0/R \quad\hbox{where}\quad \boxed{\tau= L/R}  \tag{DEI}
$$

is our time constant. If we multiply (DEI) by R and recall that $V_1=IR$ we find that $V_1$ obeys

$$
  \boxed{\tau V'_1(t) + V_1(t) = V_0(t)}   \tag{DEV}
$$

Staying close to our experiment we suppose that the input, $V_0(t)$, is sinusoidal with amplitude $A_0$ (in Volts) and frequency $f$ (in Hertz). To be precise, we take

$$
  V_0(t) = A_0\cos(2\pi ft).  \tag{V0}
$$

Our experiments suggest that the output, $V_1(t)$, has the same frequency but different amplitude and (if we had a better oscilloscope) a different phase. To be precise, we assume

$$
  V_1(t) = A_1\cos(2\pi ft - \phi)  \tag{V1}
$$

On direct substitution of formulas (V0) and (V1) into the differential equation (DEV) we arrive at the trigonometric equation

$$
  -\tau A_1 2\pi f\sin(2\pi ft - \phi) + A_1\cos(2\pi ft - \phi) = A_0\cos(2\pi ft). \tag{TE}
$$

As this equation is to hold for all time, we may pick two convenient times that lead to relatively easy deduction of the unknown amplitude, $A_1$, and phase shift, $\phi$. The first choice is a value of $t$ for which the right hand side of (TE) vanishes. As $\cos$ vanishes at $\pi/2$ we set
$$
  t = 1/(4f)
$$
in (TE) and so find
$$
  -\tau A_1 2\pi f\sin(\pi/2 - \phi) + A_1\cos(\pi/2 - \phi) = 0,
$$
that is,
$$
  -\tau 2\pi f\cos(\phi) + \sin(\phi) = 0, \quad\hbox{so}\quad \tau 2\pi f\cos(\phi) = \sin(\phi). \tag{PhiId}
$$
This determines the phase shift via
$$
  \boxed{\tan(\phi) = 2\pi f\tau.}
$$
To find $A_1$ we next $t=0$ in (TE) and find
$$
  \tau A_1 2\pi f\sin(\phi) + A_1\cos(\phi) = A_0.
$$
Using (PhiId) this takes the form
$$
  A_1\cos(\phi)(1+(2\pi f\tau)^2)) = A_0
$$
and so
$$
  A_1 = \frac{A_0}{\cos(\phi)(1+(2\pi f\tau)^2))}.
$$
We reduce the $\cos(\phi)$ term by returning to (PhiId)

$$
  (2\pi f\tau)^2\cos^2(\phi) = \sin^2(\phi) = 1 - \cos^2(\phi)  \quad\hbox{so}\quad \cos^2(\phi) = \frac{1}{(2\pi f\tau)^2+1}
$$
and so
$$
  \boxed{A_1(f) = \frac{A_0}{\sqrt{(2\pi f\tau)^2+1}}} \tag{A1}
$$

With $A_0=1\ V$ per our experiment, we now graph this $A_1$ over our range of frequencies, and compare the curve to our experimental findings.


We turn to python to graph this function, and compare it to our tinkercad work.

```python
plt.semilogx(freq, VoutPP, 'ko-')  # experiment

R = 10 # Ohm
L = 100e-3  # Henrys
tau = L/R
NewFreq = np.logspace(-1,4,100)   # 100 points between 10**(-1) and 10**4
A1 = 1/np.sqrt((2*np.pi*tau*NewFreq)**2 + 1)  # theory
plt.semilogx(NewFreq, A1, 'r')

plt.legend(['Experiment','Theory'])

plt.grid('on')
plt.xlabel('frequency, Hz')
plt.ylabel('VoutPP, Volts')
plt.title('Figure 2. LR Filter, $R=10\,\Omega$, $L=100\,mH$')
```
