# Solenoids & Inductors

A solenoid is a coil of wire. As current passes through the wire a magnetic field is created. This field is confined to the interior of coil, directed along the axis of the coil and its strength is proportional to both the current and the number of turns (coils) per unit length.

These conclusions stem from 

1. our previous calculation of the field produced, on axis, from a single loop, 
2. our presumption that the length of the coil greatly exceeds its diameter, and
3. Ampere's Law

With regard to Fig. 12.19 in our text, this first point reads

$$
  \vec B(0,y,0) = (0,\mu_0 nI, 0)
$$

for any point, $(0,y,0)$, along the coil's axis. Here $I$ denotes current, and $n$ is the number of turns per unit length.

Next, our second assumption, of a long thin coil, allows us to predict that $\vec B(x,y,z)$ runs parallel to the $y$-axis at all other points, $(x,y,z)$, - for there are an infinite number of small current segments that conspire to straighten out the field.

From there, one can construct a rectangle over which to consider the circulation of $\vec B$ and so infer its magnitude off axis.

In particular, choosing the rectangle in Fig. 12.20 in our text, with sides $S_1$, $S_2$, $S_3$ and $S_4$, Ampere's Law delivers

$$
  \oint_{S_1\cup S_2\cup S_3\cup S_4}\vec B\cdot\vec t\,ds = \mu_0 n\ell I    \tag{Alaw}
$$

where $\ell$ is the length of $S_1$. As the rectangle is easily divided into four sides we write the circulation as

$$
  \oint_{S_1\cup S_2\cup S_3\cup S_4}\vec B\cdot\vec t\,ds = \int_{S_1}\vec B\cdot\vec t\,ds +
    \int_{S_2}\vec B\cdot\vec t\,ds + \int_{S_3}\vec B\cdot\vec t\,ds + \int_{S_4}\vec B\cdot\vec t\,ds
$$

As $\vec B$ is presumed parallel to $(0,1,0)$ everywhere, while $\vec t$ is parallel to $(1,0,0)$ on sides $S_2$ and $S_4$,
the dot products, $\vec B\cdot\vec t$, vanish on those two sides and the integrals contribute nothing to the circulation.

Next, as $\vec t$ is parallel to $\vec B$ on $S_1$, and $\vec B$ has a constant field strength, $B$, there

$$
  \int_{S_1}\vec B\cdot\vec t\,ds = B\ell   \tag{S1}
$$

Finally, we suppose the $S_3$ side to be so far from the coil that the B-field is practically zero there. As such, we simply reconcile (Alaw) and (S1) and conclude that

$$
  \vec B(x,y,z) = \cases{\mu_0nI(0,1,0), & for $(x,y,z)$ inside the coil. \\ 0, & for $(x,y,z)$ outside the coil.\\}
$$

## Faraday's Law

As the flux of a magnetic field, $\vec B\cdot\vec n$, through a surface, $S$, over time it generates a ciculation, $\vec E\cdot\vec t$, in the associated electric field around, $\partial S$, the boundary of $S$. This is very tersely stated as

$$
  \boxed{\oint_{\partial S} \vec E\cdot\vec t = - \frac{d}{dt}\int\int_S \vec B\cdot\vec n} \tag{FL}
$$

where $\vec n$ is the **normal** to the surface $S$. Please learn to say this as **the change in the magnetic flux equals the circulation of the electric field.**

In general this requires real skill in manipulating surfaces, curves, tangents, normals, and fields. The beauty of the solenoid is that the left side of (FL) reduces to a potential difference (using $\vec E=-\nabla V$ and the Fundamental Theorem of Calculus)

$$
  \oint_{\partial S} \vec E\cdot\vec t = -\oint_{\partial S} \nabla V \cdot \vec t =
    -\int_a^b \left(\frac{\partial V}{\partial x}x'(s) + \frac{\partial V}{\partial y}y'(s) + 
       \frac{\partial V}{\partial z}z'(s)\right)\,ds = V(a)-V(b) 
$$

while, with $S$ denoting the cross section (of area $A$) of $N$ turns of a solenoid of length $\ell$, and the fact that $\vec B$ is parallel to $\vec n$ and scales like $\mu_0(N/\ell)I$ in a solenoid, we find

$$
  \int\int_S \vec B\cdot\vec n = \mu_0 (N/\ell)N A I
$$

On reconciling these terms we arrive at the **inductor law**

$$
  \boxed{\Delta V = L I'} \qquad \hbox{where} \qquad L = \mu_0 N^2A/\ell
$$

is the **inductance** of the solenoid, and $\Delta V = V_{head}-V_{tail}$ where head and tail are defined by the arrow used to mark the current, $I$.

On comparing this law to the our two other device laws,

$$
  \Delta V = IR  \qquad\hbox{and}\qquad  I = C(\Delta V)'
$$

our inductor law "is ordered" more like the resistor law, and so we should not be surprised that it combines, in series and parallel circuits, like the resistor.
