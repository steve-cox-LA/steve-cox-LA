# The Capacitor

With regard to the two sheets of opposite charge that concluded our previous notebook we suppose that each sheet has area $A$ and are separated by a distance $d$.

The top (red) sheet has positive charge, $Q=\sigma A$, where $\sigma>0$, and so the field lines point away, while the bottom (blue) sheet has negative charge, $-Q = -\sigma A$, and so the field lines point in. 

Between the sheets, the field lines point in the same direction and so the resulting field is doubled, while above the red and below the blue to the field lines point in opposite directions and so cancel one another. Hence
$$
  \boxed{\vec E_{cap}(z) = \cases{(0,0,-4\pi k_e\sigma) &if 0<z<d \\ (0,0,0) & otherwise}}
$$
where cap is short for capacitor, the name of this object.

In general, the **vector** Electric Field is the gradient of a **scalar** Voltage. That is
$$
  \vec E(x,y,z) = -\left(\frac{\partial V(x,y,z)}{\partial x}, \frac{\partial V(x,y,z)}{\partial y}, \frac{\partial V(x,y,z)}{\partial z}\right)
$$
that is, the $x$-component of the Electric field is the negative of the change in voltage with respect to $x$, and similarly for $y$ and $z$. From this relationship we learn that dimension of Voltage is $N\, m/C$. As you recall, a Newton meter is a Joule, and so a **Volt is a Joule per Coulomb**.

With respect to our capacitor we find the voltage (between the sheets) to be
$$
  V_{cap}(z) = 4\pi k_e\sigma z, \quad Volts
$$
This is $0$ on the bottom sheet and $4\pi k_e\sigma d$ on the top sheet. We find
$$
  V_{cap}(d) - V_{cap}(0) = 4\pi k_e\sigma d = 4\pi k_e Q d/A
$$
is typically rearranged in the form of the **Static Capacitor Law**
$$
  \boxed{C\Delta V = Q, \qquad  C \equiv \frac{A}{4\pi k_e d}}
$$
where **C is the capacitance** and $\Delta V$ is the **voltage drop**. We see that capacitance is measured in units of charge per volt, i.e., charge squared per Joule, or, in honor of Michael Faraday, a **Farad**, or F for short.

The Static Capacitor Law become Dynamic (and way more useful) when we allow the charge $Q$ to change with time. For **current** is precisely the change in charge with respect to time:
$$
  I = \frac{\Delta Q}{\Delta t} = Q'(t)
$$
which brings up the new unit **Ampere = Coulomb per Second** or **A=C/s** for short. Taking the time derivative of our static law now yields the **Dynamic Capacitor Law**
$$
  \boxed{C(\Delta V)'(t) = I(t)}
$$
In words, "the capacitance times the time rate of change of the voltage drop across the capacitor equals the current."
