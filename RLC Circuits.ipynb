We record and graph the output voltage over frequencies near 50 Hz.

```python
import numpy as np
import matplotlib.pyplot as plt

freq = np.array([20, 30, 40, 45, 50, 55, 60, 70, 80, 90])      # Hz
VoutPP = np.array([0.14, 0.28, 0.58, 0.9, 0.95, 0.85, 0.65, 0.42, 0.31, 0.24 ])  # Volts, read from oscilloscope.

plt.plot(freq, VoutPP, 'ko-')
plt.grid('on')
plt.xlabel('frequency, Hz')
plt.ylabel('VoutPP, Volts')
plt.title('LCR Filter, $R=10\Omega$, $L=100\,mH$, $C=100\,\mu F$')
```

```python
# let's draw and label our RLC circuit
import matplotlib.pyplot as plt   
import schemdraw
import schemdraw.elements as elm
d = schemdraw.Drawing()

L = d.add(elm.Inductor, label='$L$')
d.labelI(L, label='$I$', arrowofst=0.5, arrowlen=1.5, top=False)
node = d.add(elm.DOT)
node.add_label('$V_1$',loc = 'top') 

C = d.add(elm.Capacitor, label='$C$')
d.labelI(C, label='$I$', arrowofst=0.5, arrowlen=1.5, top=False)
node = d.add(elm.DOT)
node.add_label('$V_2$',loc = 'top') 

R = d.add(elm.Resistor, label='$R$', d='down')
d.labelI(R, label='$I$', arrowofst=1, arrowlen=1.5, top=True)

d.add(elm.LINE, d='left', tox = L.start)
d.add(elm.GND)
d.add(elm.SourceSin, d='up', label='$V_0$')

node = d.add(elm.DOT)
node.add_label('$V_0$',loc = 'top') 

d.draw()
```

With this schematic we can now derive a differential equation for the output voltage, $V_2$.
Our Inductor Law reads

$$
  V_0 - V_1 = LI'  \tag{ind}
$$

Our Capacitor Law reads

$$
  I = C(V_1-V_2)' = CV'_1 - CV'_2  \tag{cap}
$$

while our Resistor Law reads

$$
  V_2 = IR \tag{res}
$$

Sunstitution of (rec) into (cap) brings

$$
  I = CV'_1 - RCI' \tag{cap2}
$$

Solving (ind) for $V_1=V_0-LI'$ we then differentiate to obtain

$$
  V'_1 = V'_0 - LI''  \tag{ind2}
$$

On substition of this into (cap2) brings

$$
  I = CV'_0 - LCI'' - RCI'
$$

or, on rearrangement

$$
  LCI'' + RCI' + I = CV'_0 \tag{DEI}
$$

and finally multiplying across by $R$ and using $V_2=IR$ we arrive at

$$
  \boxed{LCV_2'' + RCV_2' + V_2 = RCV'_0}  \tag{LCR}
$$

We continue to make our standard hypotheses: The input is

$$
  V_0(t) = A_0\cos(2\pi ft).  \tag{V0}
$$

and the output is

$$
  V_2(t) = A_2\cos(2\pi ft - \phi)  \tag{V2}
$$

On direct substitution of formulas (V0) and (V2) into the differential equation (LCR) we arrive at the trigonometric equation

$$
  -LC A_2 (2\pi f)^2\cos(2\pi ft - \phi) - RCA_22\pi f\sin(2\pi ft - \phi) 
    + A_2\cos(2\pi ft - \phi) = -RCA_02\pi f\sin(2\pi ft). \tag{TE}
$$

As this equation is to hold for all time, we may pick two convenient times that lead to relatively easy deduction of the unknown amplitude, $A_2$, and phase shift, $\phi$. The first choice is a value of $t=0$ for which the right hand side of (TE) vanishes. This choice in
in (TE) brings

$$
  -LC A_2 (2\pi f)^2\cos(\phi) + RCA_22\pi f\sin(\phi) 
    + A_2\cos(\phi) = 0.
$$

On canceling $A_2$ and combining like terms we find

$$
  (LC(2\pi f)^2-1)\cos(\phi) = RC2\pi f\sin(\phi)   \tag{phi}
$$

so

$$
  \tan(\phi) = \frac{LC(2\pi f)^2-1}{RC2\pi f}
$$

Next, we choose $t=1/(4f)$ in (TE) and find

$$
  -LC A_2 (2\pi f)^2\cos(\pi/2 - \phi) - RCA_22\pi f\sin(\pi/2 - \phi) 
    + A_2\cos(\pi/2 - \phi) = -RCA_02\pi f.
$$

or

$$
 -LC A_2 (2\pi f)^2\sin(\phi) - RCA_22\pi f\cos(\phi) 
    + A_2\sin(\phi) = -RCA_02\pi f.
$$ 

collecting terms

$$
 A_2(LC(2\pi f)^2-1)\sin(\phi) + RCA_22\pi f\cos(\phi) 
     = RCA_02\pi f.
$$ 

Now using (phi) to sub for $\sin(\phi)$

$$
 A_2(LC(2\pi f)^2-1)^2\cos(\phi) + (RC)^2A_2(2\pi f)^2\cos(\phi) 
     = (RC)^2A_0(2\pi f)^2.
$$ 

and solve for 

$$
  A_2 = \frac{(2\pi fRC)^2A_0}{\cos(\phi)((LC(2\pi f)^2-1)^2 + (2\pi fRC)^2)}
$$

To remove this $\cos(\phi)$ term we square (phi) in invoke pythagoras

$$
  (LC(2\pi f)^2-1)^2\cos^2(\phi) = (RC2\pi f)^2\sin^2(\phi) = (2\pi f RC)^2(1-\cos^2(\phi))
$$

and find

$$
  \cos^2(\phi) = \frac{(2\pi f RC)^2}{(2\pi f RC)^2 + (LC(2\pi f)^2-1)^2}
$$

and so

$$
  \boxed{A_2 = \frac{2\pi fRC}{\sqrt{(LC(2\pi f)^2-1)^2 + (2\pi fRC)^2}}A_0}
$$

With $A_0=1\ V$ per our experiment, we now graph this $A_2$ over our range of frequencies, and compare the curve to our experimental findings.

```python
f = np.linspace(10,100,200)

R = 10
L = 100e-3
C = 100e-6

fr = 1/(2*np.pi*np.sqrt(L*C))
print("fr = ", fr)

a = 2*np.pi*f*R*C
b = L*C*(2*np.pi*f)**2

A2 = a / np.sqrt((1-b)**2 + a**2)

plt.plot(f, A2, 'r')

freq = np.array([20, 30, 40, 45, 50, 55, 60, 70, 80, 90])      # Hz
VoutPP = np.array([0.14, 0.28, 0.58, 0.9, 0.95, 0.85, 0.65, 0.42, 0.31, 0.24 ])  # Volts, read from oscilloscope.

plt.plot(freq, VoutPP, 'ko-')
plt.grid('on')
plt.xlabel('frequency, Hz')
plt.ylabel('VoutPP, Volts')
plt.legend(['theory','experiment'])
plt.title('Bandpass Filter, $R=10\Omega$, $L=100\,mH$, $C=100\,\mu F$')
```

We see that $A_2$ peaks when $LC(2\pi f)^2=1$, i.e., when 

$$
  f = f_r = \frac{1}{2\pi\sqrt{LC}}
$$

and call $f_r$ the **resonant frequency**.
