As in our previous notebook we now seek a quantitative understanding of this charge/discharge phenomena.

To that end we draw a careful schematic, (re)derive the Differential Equation for the output voltage, and solve it given our knowledge of the input voltage.


```python
import matplotlib.pyplot as plt   
import schemdraw
import schemdraw.elements as elm
d = schemdraw.Drawing()

R = d.add(elm.RES, label='$R$')
d.labelI(R, label='$I$', arrowofst=0.5, arrowlen=1.5, top=False)
node = d.add(elm.DOT)
node.add_label('$V_1$',loc = 'top') 

C = d.add(elm.CAP, label='$C$', d='down')
d.labelI(C, label='$I$', arrowofst=1, arrowlen=1.5, top=True)

d.add(elm.LINE, d='left', tox = R.start)
d.add(elm.GND)
d.add(elm.SOURCE_V, d='up', label='$V_0$')

node = d.add(elm.DOT)
node.add_label('$V_0$',loc = 'top') 

d.draw()
```

Ohm's Law delivers
$$
  V_0-V_1 = RI
$$
while the Capacitor Law provides
$$
  I = CV'_1
$$
Substitution of the second into the first reveals
$$
  V_0 - V_1 = RC V'_1.
$$
We rearrange to find

$$
  \boxed{\tau V'_1 + V_1 = V_0} \quad\hbox{where}\quad \boxed{\tau= RC}  \tag{DE}
$$

is our time constant. This is indeed the same DE we studied in our previous notebook. The difference here is that we will focus on the charging phase of our experiment. To that end, we suppose that our output voltage, $V_1$, starts from 0, i.e.,

$$
  V_1(0) = 0,
$$

at which time $V_0$ turns on (or goes high), which in the Arduino world is $V_0=5$. Our (DE) then take the concrete form

$$
  \tau V'_1(t) + V_1(t) = 5, \qquad V_1(0)=0   \tag{DE5}
$$

We simplify this somewhat by writing $V_1$ as a constant part minus a dynamic part, namely

$$
   V_1(t) = 5 - u(t)
$$

When we plug this guess into (DE5) we arrive at a DE for u,

$$
  \tau u'(t) + u(t) = 0, \qquad u(0) = 5   \tag{DEu}
$$

Now this poses the question of what function looks like a multiple of its derivative. We will soon see in ENGR 1122 that the exponential function fits the bill. In particular,

$$
  (\exp(t))' = \exp(t)
$$

and so, via the chain rule,

$$
  (\exp(-t/\tau))' = -(1/\tau)\exp(-t/\tau)
$$

hence, $u(t)=5\exp(-t/\tau)$, solves (DEu) and so


$$
  \boxed{V_1(t) = 5(1-\exp(-t/\tau))}   \tag{V1}
$$

solves (DE5). We turn to python to graph this function, and compare it to the charging phase of our tinkercad work.




```python
import matplotlib.pyplot as plt
import numpy as np

R = 10e3
C = 1e-6
tau = R*C

V0 = 5
t = np.linspace(0,0.1,100)
V1 = V0 * (1-np.exp(-t/tau))

plt.plot(t*1e3, V1)
#plt.plot([0, 5], [0.632*V0, 0.632*V0], 'r')
plt.xlabel('t  (milliseconds)')
plt.ylabel('$V_1$  (volts)')
#plt.legend(['$V_1$', '$0.632*V_0$'])
plt.grid('on')
plt.title('Capacitor Charging with $\\tau$ = %d  (ms)' %(round(tau*1e3)))
```

That explicit formula, (V1), for $V_1$ gives us ready means to build a capacitance meter. We note at time $t=\tau$ that

$$
  V_1(\tau) = 5(1-\exp(-1)) = 5\cdot 0.632
$$

and so, if we switch our pulse on, keep track of time and run time until $V_1=5\cdot 0.632$ then that time is then $\tau$. Knowing $\tau$ and $R$ we simply
solve for

$$
  C = \tau/R
$$

To build this device with an Arduino, we recall that 5 volts corresponds to a numerical value of 1023, and so we should switch our input from high to low and proceed until $V_1>672$ as $1023*\cdot 0.632=672$. This 672 threshold runs too high in practice (at least in the tinkercad cloud). And so I have reduced it to 648 in the simulation below.

This example was using a capacitance of 100 $\mu F$. Of course we must explore its performance over a range of capacitances.

```python
CapTrue = np.array([1e-2, 1e-1, 1e0, 1e1, 1e2])    # micro Farads
CapMeas = np.array([10.4e-3, 101.2e-3, 1.01, 10.07, 100.75])   # using 648 for V1 threshold, and R = 10 k

RelErr = 100.0*np.abs(CapTrue - CapMeas)/CapTrue

plt.semilogx(CapTrue, RelErr, 'ko-')
plt.grid('on')
plt.xlabel('True Capacitance, $\\mu F$')
plt.ylabel('Percent Error')
plt.title('Capacitance Meter: $R=10\ k\Omega$')
```

We see that our meter does poorly on capacitances smaller than 100 nF. At issue is the smaller value of $\tau=RC$. It is likely too small for our Arduino to measure accurately. The likely solution is to increase the vale of R by a factor of 10.
