```python
# we now graph the peak-to-peak output voltage over the range of frequencies

import matplotlib.pyplot as plt   
import numpy as np

freq = np.array([1e1, 1e2, 1e3, 1e4])
VoutPP = np.array([1, 0.92, 175e-3, 16e-3])  # read from oscilloscope. Be CAREFUL to note V or mV or ...

plt.semilogx(freq, VoutPP, 'ko-')
plt.grid('on')
plt.xlabel('frequency, Hz')
plt.ylabel('VoutPP, Volts')
plt.title('Figure 1. RC Filter, $R=1\,k\Omega$, $C=1\,\mu F$')
```

As the size of the output voltage descreases with incresaing frequency we call this circuit a **Low Pass Filter**.

With these experiments under our belts we dive back into the physics in the hope of clarifying this picture.

```python
# to build the theory we need a well labeled circuit diagram

import schemdraw
import schemdraw.elements as elm
d = schemdraw.Drawing()

R = d.add(elm.RES, label='$R$')
d.labelI(R, label='$I$', arrowofst=0.5, arrowlen=1.5, top=False)
node = d.add(elm.DOT)
node.add_label('$V_1$',loc = 'top') 

C = d.add(elm.CAP, label='$C$', d='down')
d.labelI(C, label='$I$', arrowofst=1, arrowlen=1.5, top=True)

d.add(elm.LINE, d='left', tox = R.start)
d.add(elm.GND)
d.add(elm.SourceSin, d='up', label='$V_0$')

node = d.add(elm.DOT)
node.add_label('$V_0$',loc = 'top') 

d.draw()
```

Do you see the correspondence with our experimental work? The function generator is there in $V_0$, while the oscilloscope measures $V_1$. We will use our two device laws to arrive at a differential equation for $V_1$.

Ohm's Law delivers
$$
  V_0-V_1 = RI
$$
while the Capacitor Law provides
$$
  I = CV'_1
$$
Substitution of the second into the first reveals
$$
  V_0 - V_1 = RC V'_1.
$$
We rearrange this and arrive at the **Differential Equation**
$$
  \boxed{\tau V'_1(t) + V_1(t) = V_0(t)} \tag{DE}
$$
where
$$
  \boxed{\tau= RC}
$$
is our circuit **time constant**. Can you confirm that $\tau$ has units of seconds? For the $R$ and $C$ used in our experiments note that $\tau = 1\, ms$.

Staying close to our experiment we suppose that the input, $V_0(t)$, is sinusoidal with amplitude $A_0$ (in Volts) and frequency $f$ (in Hertz). To be precise, we take

$$
  V_0(t) = A_0\cos(2\pi ft).  \tag{V0}
$$

Our experiments suggest that the output, $V_1(t)$, has the same frequency but different amplitude and (if we had a better oscilloscope) and a different phase. To be precise, we assume

$$
  V_1(t) = A_1\cos(2\pi ft - \phi)  \tag{V1}
$$

On direct substitution of formulas (V0) and (V1) into the differential equation (V1DE) we arrive at the trigonometric equation

$$
  -\tau A_1 2\pi f\sin(2\pi ft - \phi) + A_1\cos(2\pi ft - \phi) = A_0\cos(2\pi ft). \tag{TE}
$$

As this equation is to hold for all time, we may pick two convenient times that lead to realatively easy desuction of the unknown amplitude, $A_1$, and phase shift, $\phi$. The first choice is a value of $t$ for which the right hand side of (TE) vanishes. As $\cos$ vanishes at $\pi/2$ we set
$$
  t = 1/(4f)
$$
in (TE) and so find
$$
  -\tau A_1 2\pi f\sin(\pi/2 - \phi) + A_1\cos(\pi/2 - \phi) = 0,
$$
that is,
$$
  -\tau 2\pi f\cos(\phi) + \sin(\phi) = 0, \quad\hbox{so}\quad \tau 2\pi f\cos(\phi) = \sin(\phi). \tag(PhiId)
$$
This determines the phase shift via
$$
  \boxed{\tan(\phi) = 2\pi f\tau.}
$$
To find $A_1$ we next $t=0$ in (TE) and find
$$
  \tau A_1 2\pi f\sin(\phi) + A_1\cos(\phi) = A_0.
$$
Using (PhiId) this takes the form
$$
  A_1\cos(\phi)(1+(2\pi f\tau)^2)) = A_0
$$
and so
$$
  A_1 = \frac{A_0}{\cos(\phi)(1+(2\pi f\tau)^2))}.
$$
We reduce the $\cos(\phi)$ term by returning to (PhiId)

$$
  (2\pi f\tau)\cos^2(\phi) = \sin^2(\phi) = 1 - \cos^2(\phi)  \quad\hbox{so}\quad \cos^2(\phi) = \frac{1}{(2\pi f\tau)^2+1}
$$
and so
$$
  \boxed{A_1(f) = \frac{A_0}{\sqrt{(2\pi f\tau)^2+1}}}
$$

With $A_0=1\ V$ per our experiment, we now graph this $A_1$ over our range of frequencies, and compare the curve to our experimental findings.

```python
plt.semilogx(freq, VoutPP, 'ko-')  # experiment

tau = 1e-3
NewFreq = np.logspace(0,5,100)   # 100 points between 10**0 and 10**5
A1 = 1/np.sqrt((2*np.pi*tau*NewFreq)**2 + 1)  # theory
plt.semilogx(NewFreq, A1, 'r')

plt.legend(['Experiment','Theory'])

plt.grid('on')
plt.xlabel('frequency, Hz')
plt.ylabel('VoutPP, Volts')
plt.title('Figure 2. RC Filter, $R=1\,k\Omega$, $C=1\,\mu F$')

```

It is more than just "nice" that theory and experiment correspond. For we shall soon see that theory will open the door to a means for actually measuring capacitance.  

To get there we first need a more common measure of the **Gain** of a device. Gain is measured in **decibels** (10 bels) where a **bel** is the base-10 log of the ratio of output power to input power. As power is proportional to the square of voltage, we find the Gain of our circuit to be

$$
  {\rm Gain}(f) = 10\log_{10}\left(\frac{A_1^2(f)}{A_0^2}\right) 
          = 10\log_{10}\frac{1}{1+(2\pi f\tau)^2} 
          = -10\log_{10}(1+(2\pi f\tau)^2)
$$

where in the last step we used the facts (to be derived in ENGR 1122) that the log(a/b) = log(a) - log(b) and log(1) = 0.

```python
Gain = -10*np.log10((2*np.pi*tau*NewFreq)**2 + 1)  
plt.semilogx(NewFreq, Gain, 'r')

plt.grid('on')
plt.xlabel('frequency, Hz')
plt.ylabel('Gain, dB')
plt.title('Figure 3. RC Filter, $R=1\,k\Omega$, $C=1\,\mu F$')
```

We see that Gain is 0 when the output amplitude matches the input amplitude. The **cutoff** frequency, $f_{cut}$, of a filter is defined as the frequency at which the output power is one half of the input power. As

$$
  10\log_{10}(1/2) = -3
$$

we see that $f_{cut}$ is the frequency at which the Gain is -3 dB. Hence we solve 
$$
  -3 = -10\log_{10}(1+(2\pi f_{cut}\tau)^2),
$$
via
$$
  (3/10) = \log_{10}(1+(2\pi f_{cut}\tau)^2), \quad\hbox{so}\quad 10^{3/10} = 1+(2\pi f_{cut}\tau)^2
$$
so
$$
  1 = 2\pi f_{cut}\tau
$$
and finally,
$$
  \boxed{f_{cut} = \frac{1}{2\pi\tau}}
$$
In our example this yields 
$$
  f_{cut} = 159.15\ Hz
$$
We illustrate this by zooming into our previous Gain plot.

```python
plt.semilogx(NewFreq, Gain, 'r')
plt.semilogx(NewFreq, -3*np.ones(Gain.shape),'k')
plt.text(2, -3.6, 'Gain = -3db', color='k')

fcut = 1/(2*np.pi*tau)
plt.semilogx([fcut, fcut],[-10, 1],'b')
plt.text(100, -9, '$frequency = f_{cut}$', color='b', rotation=90)

plt.xlim([1,1e4])
plt.ylim([-10,1])
plt.grid('on')
plt.xlabel('frequency, Hz')
plt.ylabel('Gain, dB')
plt.title('Figure 4. RC Filter, $R=1\,k\Omega$, $C=1\,\mu F$')
```

So, here at last is the idea for a **capacitance measurement device**. Given an unknown capacatance, $C$, place it in series after a known resistance, $R$, then drive it with sinusoidal input over a range of frequencies and keep track of the gain. Once you see a gain of -3 you have your $f_{cut}$ and can then use

$$
  f_{cut} = \frac{1}{2\pi RC}
$$
to solve for the unknown
$$
  \boxed{C = \frac{1}{2\pi Rf_{cut}}}
$$
